**<font color='darkblue'>Nobel Prizes</font>**

*The Nobel Prizes are a set of prestigious international awards bestowed annually in several categories, including Physics, Chemistry, Medicine or Physiology, Literature, Peace, and Economic Sciences. They were established by the will of Alfred Nobel, a Swedish inventor, engineer, and industrialist, who left his entire estate to fund the prizes upon his death in 1896*                           
 
**<font color='darkblue'>Nobel Prize Winners Dataset** 

The Nobel Prize Winners dataset is a collection of data about the recipients of the Nobel Prize, one of the most prestigious awards given for outstanding achievements in physics, chemistry, medicine, literature, peace, and economic sciences. This dataset typically includes information such as:     

     

- **<font color='DarkBlue'>Year**:The year in which the Nobel Prize was awarded.  
- **<font color='DarkBlue'>Category**: The category of the Nobel Prize (Physics, Chemistry, Medicine, Literature, Peace, Economic Sciences)  
- **<font color='DarkBlue'>Prize** :  The Prize indicates the specific achievement or contribution for which the laureate and year was awarded the Nobel Prize.  
- **<font color='DarkBlue'>Motivation**: The reason or motivation for the Nobel Prize award.  
- **<font color='DarkBlue'>Prize Share**: For awards shared by multiple laureates, this indicates the portion of the prize each recipient received.   
- **<font color='DarkBlue'>Laureate ID**: This column contains a unique identifier assigned to each Nobel laureate. It's used to distinguish between different individuals who have won the Nobel Prize, particularly useful when multiple people share a prize.     
- **<font color='DarkBlue'>Laureate Type**: This column contains the type of recipient or laureate such as indivudual,organization or institution
- **<font color='DarkBlue'>Full Name**: The full name of the Nobel laureate.  
- **<font color='DarkBlue'>Birth Date**: The date of birth of the laureate.  
- **<font color='DarkBlue'>Birth City,Country**: The place (city, country) where the laureate was born.  
- **<font color='DarkBlue'>Sex**: The gender of the laureate (male, female, or non-binary).  
- **<font color='DarkBlue'>Organization Name**: The name of the organization which the laureate is associated with.  
- **<font color='DarkBlue'>Organization City,Country**:Indicates the country and city where the organization is located.  
- **<font color='DarkBlue'>Death Date**: The date of death of the laureate.  
- **<font color='DarkBlue'>Death City & Country**: The place (city, country) where the laureate died.

```python
# The function calculates percantage of all null values in data:
#Verideki toplam null veri sayƒ±sƒ±nƒ± ve bunun y√ºzdeliƒüinin sorgulanmasƒ±nƒ± saƒülayan fonksiyon:
def missing_values(df):
    missing_number = df.isnull().sum().sort_values(ascending=False)
    missing_percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending=False)
    missing_values = pd.concat([missing_number, missing_percent], axis=1, keys=['Missing_Number', 'Missing_Percent'])
    return missing_values[missing_values['Missing_Number']>0]
```

```python
#The Country has won the most nobel prizes throughout the years and total count of prizes by years:
total_prizes=dfm[dfm['birth_country']=='United States of America'].groupby(['year']).agg({'prize':'count'}).sort_values('year',ascending=True).reset_index()
total_prizes.head(15)
```

```python
#The Country has won the most nobel prizes throughout the years and count of prizes by years in various categories:
count_nobelprizes=dfm[dfm['birth_country']=='United States of America'].groupby(['year','category']).agg({'prize':'count'}).sort_values('year',ascending=True).reset_index()
count_nobelprizes.columns=['year','category','countofprizes']
count_nobelprizes.head(20)
```

```python
# Filtering the data for the first three countries that have won the most Nobel Prizes :
dfc=dfm.groupby(['year','birth_country']).agg({'prize':'count'}).sort_values(by='year',ascending=False).reset_index()
dfct=dfc[dfc['birth_country'].isin(['United States of America','United Kingdom','Germany'])]
dfct
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-white') 
dfc = dfm.groupby(['year', 'birth_country']).agg({'prize': 'count'}).sort_values(by='year', ascending=False).reset_index()
dfct = dfc[dfc['birth_country'].isin(['United States of America', 'United Kingdom', 'Germany'])]

# Drawing Line Plot:
plt.figure(figsize=(15,5))
sns.lineplot(data=dfct, x='year', y='prize', hue='birth_country', marker='o')

# Title arrangements:
plt.title('Count of Nobel Prizes by Country Over Years', fontsize=16)
plt.xlabel(' ', fontsize=14)
plt.ylabel('Number of Nobel Prizes', fontsize=14)
plt.xticks(np.arange(1905, 2020, step=3),fontsize=12,rotation=90)
plt.yticks(np.arange(0, 10, step=1))
plt.legend(title='Birth Country')
plt.grid(True)
plt.tight_layout()
plt.show()

```
